## Full test data representation
Visualization of the full data by selecting Key 1 and Key 2 from the dropdown.
It uses Dash.

```python
df = pd.read_parquet("../data/test_full.parquet")
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html

app = dash.Dash()

fnameDict = {}
for i in range(len(df.index.get_level_values('key_1').unique().to_list())):
    fnameDict[df.index.get_level_values('key_1').unique().to_list()[i]] = df.loc[df.index.get_level_values('key_1').unique().to_list()[i]].index.get_level_values('key_2').unique().to_list()
names = list(fnameDict.keys())
nestedOptions = fnameDict[names[0]]

app.layout = html.Div(
    [
        html.Div([
        dcc.Dropdown(
            id='name-dropdown',
            options=[{'label':name, 'value':name} for name in names],
            value = list(fnameDict.keys())[0]
            ),
            ],style={'width': '20%', 'display': 'inline-block'}),
        html.Div([
        dcc.Dropdown(
            id='opt-dropdown'
            ),
            ],style={'width': '20%', 'display': 'inline-block'}
        ),
        html.Hr(),
        dcc.Graph(id='indicator-graphic')
    ]
)

@app.callback(
    dash.dependencies.Output('opt-dropdown', 'options'),
    [dash.dependencies.Input('name-dropdown', 'value')]
)
def update_date_dropdown(name):
    return [{'label': i, 'value': i} for i in fnameDict[name]]

@app.callback(
    Output('indicator-graphic', 'figure'),
    [dash.dependencies.Input('opt-dropdown', 'value')],[dash.dependencies.Input('name-dropdown', 'value')])
def update_graph(key_2_val,key_1_val):
    df_test = df.loc[key_1_val, key_2_val]
    f = go.FigureWidget()
    #f.add_scatter(y=df_test['combined'],x=df_test.index,mode = 'markers',name = 'combined', marker={'size': 20,'color': 3,'opacity': 0.6,'colorscale': 'Viridis'});
    f.add_scatter(y=df_test['target'],x=df_test.index,mode = 'markers',name = 'Actual Target', marker={'size': 10,'color': 5,'opacity': 0.6,'colorscale': 'Viridis'});
    f.add_scatter(y=df_test['target_mean'],x=df_test.index,mode = 'markers',name = 'Forecasted Targeted Mean', marker={'size': 10,'color': 7,'opacity': 0.6,'colorscale': 'Viridis'});
    f.layout.title = 'Forecast Visualization'  
    return f



if __name__ == '__main__':
    app.run_server()
```
